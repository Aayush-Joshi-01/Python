```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
```

```python
df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B'])
```

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
```

```python
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df = pd.DataFrame({
       "lev1": [1, 1, 1, 2, 2, 2],
       "lev2": [1, 1, 2, 1, 1, 2],
       "lev3": [1, 2, 1, 2, 1, 2],
       "lev4": [1, 2, 3, 4, 5, 6],
       "values": [0, 1, 2, 3, 4, 5]})
df
```

```python
df.pivot(index="lev1", columns=["lev2", "lev3"], values="values")
```

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc="min", fill_value=0)
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc="max", fill_value=0)
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc="sum", fill_value=0)
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc="mean", fill_value=0)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                       aggfunc={'D': "mean",
                                'E': ["min", "max", "mean"]})
table
```
