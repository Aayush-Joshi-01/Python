```python
def convert_date_columns(df, date_columns):
    for col in date_columns:
        df[col] = pd.to_datetime(df[col], format='%d-%b-%y')
    return df
```

```python
date_columns = ['Training Start Date', 'Training End Date']
df = convert_date_columns(df, date_columns)
df[['Training Start Date', 'Training End Date']]
```

```python
def convert_to_lst(row):
    print(row)
    if ',' in row:
        return row.lower().replace(" ",'').split(',')
    elif "+" in row:
        return row.lower().replace(" ",'').split('+')
    else:
        return [row.lower().replace(" ",'')]
```

```python
def extract_skills(row):
    skills = set()
    if pd.notna(row['Current Skills']):
        skills.update(convert_to_lst(row['Current Skills']))
    if pd.notna(row['Training Stack']):
        skills.update(convert_to_lst(row['Current Skills']))
    if pd.notna(row['Upgraded Skills']):
        skills.update(convert_to_lst(row['Current Skills']))
    return list(skills)

temp_skills = []
# Apply the function to each row and store unique skills in a set
unique_skills_set = set()
for index, row in df.iterrows():
    unique_skills_set.update(extract_skills(row))
    temp_skills.append(list(unique_skills_set))

df['All Skills'] = pd.Series(temp_skills)
df['All Skills']
```

```python
df['Training Stack List'] = df['Training Stack'].apply(convert_to_lst)
```

```python
df['Current Skills List'] = df['Current Skills'].apply(convert_to_lst)
```

```python
def fill_scores(df, preassessment_center, preassessment_std, final_score_center, final_score_std):
    current_skills_map = df.groupby('Current Skills')['Current Skills'].count().to_dict()
    for key, value in current_skills_map.items():
        df.loc[df['Current Skills'] == key, 'Pre Assessment score'] = list(np.random.normal(15, 3, value).astype(int))
        
    upgraded_skills_map = df.groupby('Upgraded Skills')['Upgraded Skills'].count().to_dict()
    for key, value in upgraded_skills_map.items():
        df.loc[df['Upgraded Skills'] == key, 'Final Score'] = list(np.random.normal(24, 5, value).astype(int))
        
    return df
```

```python
df=fill_scores(df,15,3,24,5)
df
```

```python
def training_effectiveness_analysis(data):
    # Calculate improvement in scores
    data['Score Improvement'] = data['Final Score'] - data['Pre Assessment score']
    
    return data[['Emp ID', 'Name', 'Pre Assessment score', 'Final Score', 'Score Improvement']]
```

```python
# Function to group employees Based upon any Criteria
def group_by_criteria_data(data, grouper_criteria):
    return data.groupby(grouper_criteria)['Name'].apply(list).to_dict()

def group_by_criteria_count(data, grouper_criteria):
    return data.groupby(grouper_criteria)['Name'].count()
```
