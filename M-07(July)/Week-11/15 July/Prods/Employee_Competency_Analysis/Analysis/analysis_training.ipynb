```python
import pandas as pd
```

```python
df = pd.read_excel("Training_CTR_Report2024.xlsx")
df
```

```python
def convert_date_columns(df, date_columns):
    for col in date_columns:
        df[col] = pd.to_datetime(df[col], format='%d-%b-%y')
    return df
```

```python
date_columns = ['Training Start Date', 'Training End Date']
df = convert_date_columns(df, date_columns)
df[['Training Start Date', 'Training End Date']]
```

```python
def convert_to_lst(row):
    print(row)
    if ',' in row:
        return row.lower().replace(" ",'').split(',')
    elif "+" in row:
        return row.lower().replace(" ",'').split('+')
    elif '(' in row:
        return row.lower().replace(" ",'').split('(')[0].split(',')
    elif '-' in row:
        return row.lower().replace(" ",'').split('-')
    elif ")" in row:
        return row.lower().replace(" ",'').split(')')[0].split(',')
    else:
        return [row.lower().replace(" ",'')]
```

```python
df['Training Stack'] = df['Training Stack'].apply(convert_to_lst)
```

```python
df['Current Skills'] = df['Current Skills'].apply(convert_to_lst)
```

```python
def extract_skills(row):
    skills = set()
    if pd.notna(row['Current Skills']):
        skills.update(row['Current Skills'].split(','))
    if pd.notna(row['Training Stack']):
        skills.update(row['Training Stack'].split(','))
    if pd.notna(row['Upgraded Skills']):
        skills.update(row['Upgraded Skills'].split(','))
    return list(skills)

temp_skills = []
# Apply the function to each row and store unique skills in a set
unique_skills_set = set()
for index, row in df.iterrows():
    unique_skills_set.update(extract_skills(row))
    temp_skills.append(list(unique_skills_set))

df['All Skills'] = pd.Series(temp_skills)
df['All Skills']
```

```python
df.to_csv('cleaned_data.csv')
```
