```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv')
df.head()
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.head()
```

```python
def group_by_criteria_data(data, grouper_criteria):
    return data.groupby(grouper_criteria)['Name'].apply(list).to_dict()

def group_by_criteria_count(data, grouper_criteria):
    return data.groupby(grouper_criteria)['Name'].count()
```

```python
data_grouped_by_criteria = group_by_criteria_data(df, 'All Skills')
```

```python
list(data_grouped_by_criteria.items())
```

```python
skills = ['java', 'aws', '.net', 'devops']
```

```python
list_of_employee = []
for sk in skills:
    if sk in data_grouped_by_criteria.keys():
        print(sk)
        print(sk in data_grouped_by_criteria.keys())
        list_of_employee.append
    print(sk)
```
